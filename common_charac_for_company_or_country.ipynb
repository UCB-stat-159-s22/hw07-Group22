```python
# import packages
import pandas as pd
import numpy as np
from collections import Counter
import re
import matplotlib.pyplot as plt
```

```python
chocolate = pd.read_csv('data/chocolate.csv')
chocolate
```

```python
def country_common_charac(chocolate_df, country_name):
    chocolate_country = chocolate_df[chocolate_df['Country of Bean Origin'] == country_name][["Most Memorable Characteristics"]]
    charac_lst = chocolate_country['Most Memorable Characteristics'].tolist()
    lst = []
    for i in np.arange(len(charac_lst)):
        charac_lst[i] = charac_lst[i].split(', ')

    for i in charac_lst:
        for j in i:
            lst.append(j)

    dic = Counter(lst)
    df = pd.DataFrame(dic.items(), columns=['characteristics', 'frequency']).sort_values(by=['frequency'], ascending=False)
    df['percentage'] = [i / sum(df['frequency'].tolist()) for i in df['frequency'].tolist()]
    return df
```

```python
def company_common_charac(chocolate_df, company_name):
    chocolate_country = chocolate_df[chocolate_df['Company (Manufacturer)'] == company_name][["Most Memorable Characteristics"]]
    charac_lst = chocolate_country['Most Memorable Characteristics'].tolist()
    lst = []
    for i in np.arange(len(charac_lst)):
        charac_lst[i] = charac_lst[i].split(', ')

    for i in charac_lst:
        for j in i:
            lst.append(j)

    dic = Counter(lst)
    df = pd.DataFrame(dic.items(), columns=['characteristics', 'frequency']).sort_values(by=['frequency'], ascending=False)
    df['percentage'] = [i / sum(df['frequency'].tolist()) for i in df['frequency'].tolist()]
    return df
```

```python
country_lst = []
for country in list(set(chocolate['Country of Bean Origin'])):
    df = country_common_charac(chocolate, country)
    if sum(df['frequency'].tolist()) >= 20:
        if df['percentage'][0] >= 0.1:
            country_lst.append(country)
country_lst
```

```python
company_lst = []
for company in list(set(chocolate['Company (Manufacturer)'])):
    df = company_common_charac(chocolate, company)
    if sum(df['frequency'].tolist()) >= 20:
        if df['percentage'][0] >= 0.1:
            company_lst.append(company)
company_lst
```
