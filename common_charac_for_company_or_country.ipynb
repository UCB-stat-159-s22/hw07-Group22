### Analysis of testing if a country of bean origin has a standout characteristics

```python
# import packages
import pandas as pd
import numpy as np
from collections import Counter
import re
import matplotlib.pyplot as plt
from IPython.display import display_html

from functions.utils import *
```

```python
# load data
chocolate = pd.read_csv('data/chocolate.csv')
chocolate
```

The function 'country_common_charac' returns a dataframe containing all the mentioned characteristics for a specific country along with the frequency and the percentage of the appearance of each characteristics. Below, we calculate if any country has a characteristic that appeared more than 10% of times among all the characteristics.

```python
country_lst = []
for country in list(set(chocolate['Country of Bean Origin'])):
    df = country_common_charac(chocolate, country)
    if sum(df['frequency'].tolist()) >= 20:
        if df['percentage'][0] >= 0.1:
            country_lst.append(country)
country_lst
```

To better visualize the standout characteristics of the country listed above, we print out a table for each country.

```python
Vanuatu = country_common_charac(chocolate, "Vanuatu")
Malaysia = country_common_charac(chocolate, "Malaysia")
Papua = country_common_charac(chocolate, "Papua New Guinea")
Vanuatu.to_csv('tables/Vanuatu_charac.csv')
Malaysia.to_csv('tables/Malaysia_charac.csv')
Papua.to_csv('tables/Papua_charac.csv')
```
